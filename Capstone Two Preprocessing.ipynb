```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn import preprocessing

DailyDf = pd.read_csv('/Users/josevans/Library/Mobile Documents/com~apple~CloudDocs/Python files/BTCDailyEngineered.csv', index_col='date', parse_dates=True)
DailyDf.head()
```

```python
cols_to_drop = ['close', 'high', 'low', 'open']+[col for col in DailyDf.columns if 'later' in col]
```

```python
cols_to_drop
```

```python
TrainDf = DailyDf.drop(cols_to_drop, axis=1).dropna()
```

```python
TrainDf.head()
```

```python
TrainDf.dtypes
```

```python
col_names = TrainDf.columns

cols_to_be_scaled = [col for col in TrainDf.columns if ((TrainDf[col].min() != 0) and (TrainDf[col].max() != 1))]

scaler = preprocessing.StandardScaler()

ScaledDf = scaler.fit_transform(TrainDf[cols_to_be_scaled])

ScaledDf = pd.DataFrame(ScaledDf, columns=cols_to_be_scaled, index=TrainDf.index)

newdf = ScaledDf.merge(TrainDf[TrainDf.columns.difference(cols_to_be_scaled)], on=TrainDf.index).set_index('key_0')

newdf.head()
```

```python
from sklearn.model_selection import train_test_split

X = newdf
y = DailyDf.dropna()['%chg_1_days_later']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```
